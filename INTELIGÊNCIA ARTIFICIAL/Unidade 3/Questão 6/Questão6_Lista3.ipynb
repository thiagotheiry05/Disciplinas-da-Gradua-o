```python
# Importando as bibliotecas
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque
```

```python
# Definindo a função da busca em largura
def bfs(grafico, comeco, destino):
    visitado = set()
    fila = deque([(comeco, [])])

    while fila:
        no, caminho = fila.popleft()
        if no == destino:
            return caminho + [no]

        if no not in visitado:
            visitado.add(no)
            for vizinho in grafo[no]:
                fila.append((vizinho, caminho + [no]))

    return None
```

```python
# Criando o grafo
grafo = nx.Graph()
grafo.add_edges_from([
    ('A', 'B', {'weight': 4}),
    ('A', 'C', {'weight': 3}),
    ('B', 'C', {'weight': 1}),
    ('B', 'D', {'weight': 5}),
    ('C', 'D', {'weight': 8}),
    ('C', 'E', {'weight': 10}),
    ('D', 'E', {'weight': 2}),
    ('D', 'F', {'weight': 6}),
    ('E', 'F', {'weight': 4})
])
```

```python
# Visualização do grafo
pos = nx.spring_layout(grafo)
labels = nx.get_edge_attributes(grafo, 'weight')
nx.draw_networkx(grafo, pos, with_labels=True, node_color='lightblue')
nx.draw_networkx_edge_labels(grafo, pos, edge_labels=labels)
plt.show()
```

```python
# Chamando o algoritmo de busca do menor caminho
no_comeco = 'A'
no_final = 'F'
caminho_mais_curto = bfs(grafo, no_comeco, no_final)
```

```python
# Imprimindo o menor caminho encontrado
if caminho_mais_curto:
    print(f"Menor caminho de {no_comeco} para {no_final}: {' -> '.join(caminho_mais_curto)}")
else:
    print(f"Não existe caminho entre {no_comeco} e {no_final} no grafo.")
```
