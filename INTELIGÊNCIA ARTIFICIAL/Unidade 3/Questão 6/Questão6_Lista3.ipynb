```python
# Importando as bibliotecas
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque
```

```python
# Definindo a função da busca em largura
def bfs_shortest_path(graph, start, goal):
    visited = set()
    queue = deque([(start, [])])

    while queue:
        node, path = queue.popleft()
        if node == goal:
            return path + [node]

        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                queue.append((neighbor, path + [node]))

    return None
```

```python
# Criando o grafo
graph = nx.Graph()
graph.add_edges_from([
    ('A', 'B', {'weight': 4}),
    ('A', 'C', {'weight': 3}),
    ('B', 'C', {'weight': 1}),
    ('B', 'D', {'weight': 5}),
    ('C', 'D', {'weight': 8}),
    ('C', 'E', {'weight': 10}),
    ('D', 'E', {'weight': 2}),
    ('D', 'F', {'weight': 6}),
    ('E', 'F', {'weight': 4})
])
```

```python
# Visualização do grafo
pos = nx.spring_layout(graph)
labels = nx.get_edge_attributes(graph, 'weight')
nx.draw_networkx(graph, pos, with_labels=True, node_color='lightblue')
nx.draw_networkx_edge_labels(graph, pos, edge_labels=labels)
plt.show()
```

```python
# Chamando o algoritmo de busca do menor caminho
start_node = 'A'
goal_node = 'F'
shortest_path = bfs_shortest_path(graph, start_node, goal_node)
```

```python
# Imprimindo o menor caminho encontrado
if shortest_path:
    print(f"Menor caminho de {start_node} para {goal_node}: {' -> '.join(shortest_path)}")
else:
    print(f"Não existe caminho entre {start_node} e {goal_node} no grafo.")
```
