```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando nossos dados
dataset = np.array([
    [-5.82,-4.58,-1.97],
    [-4.68 ,2.16, 3.71],
    [-4.68, 2.16 ,3.71],
    [3.36, -3.19, 1.09],
  	[7.72 ,0.88 ,1.80],
    [-7.64 ,3.06 ,3.50],
    [-6.87 ,0.57 ,-5.45],
    [4.47 ,-2.62 ,5.76],
    [7.73 ,-2.01 ,5.18],
    [-7.71 ,3.34 ,-6.33],
    [-5.91 ,-0.49 ,-5.68],
    [2.18 ,3.81 ,5.82],
    [6.72 ,-0.93 ,-3.04],
    [-5.25 ,-0.26 ,0.56],
    [-6.94 ,-1.22 ,1.13],
    [7.09 ,0.20 ,2.25],
    [6.81 ,3.17 ,-4.15],
    [-4.19 ,4.24 ,4.04],
    [-5.38 ,-1.74 ,1.43],
    [5.08 ,3.30 ,5.33],
    [7.27 ,0.93 ,-2.78]
])
```

```python
# Plotando em 3D os pontos dos nossos dados
x = dataset[:, 0]
y = dataset[:, 1]
z = dataset[:, 2]

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(x, y, z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

#Letra a) Considerar que a rede possui trs neurnio que corresponde a existncia de trs clusters nos dados. Inicializar os pesos dos neurnios (centroides) de forma aleatria

```python
num_clusters = 3
num_iterations = 100

# Inicializao aleatria dos centroides
centroids = dataset[np.random.choice(range(len(dataset)), size=num_clusters, replace=False)]

for _ in range(num_iterations):
    # Clculo das distncias entre os pontos e os centroides
    distances = np.linalg.norm(dataset[:, np.newaxis] - centroids, axis=-1)

    # Atribuio dos pontos aos clusters
    cluster_assignments = np.argmin(distances, axis=1)

    # Atualizao dos centroides
    for i in range(num_clusters):
        centroids[i] = np.mean(dataset[cluster_assignments == i], axis=0)

# Plotagem dos pontos do dataset
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(dataset[:, 0], dataset[:, 1], dataset[:, 2], c=cluster_assignments, cmap='viridis')

# Plotagem dos centroides
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2], c='red', marker='x', s=50, linewidths=2)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
